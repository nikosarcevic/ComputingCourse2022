```python
# find the mean, median, mode, and std (descriptive stats) for each distribution


# if you are repeating the same code in order to get the descriptive stats for each pdf -- automate!
# do not copy and paste code over and over again but make a function to do the work for you

def get_descriptive_stats(data, columns):
    
    """
    A wrapper function to get the descriptive statistics of a given pdf.
    This function is designed to work with a pandas dataframe.
    
    Args:
        data (pandas dataframe): the pdf (distribution)
        columns (str list): column names
        
    Returns:
        stds (list): standard deviation values
        means (list): mean values of distribution
        medians..... finish this docstring
    """
    
    
    stds = []
    means = []
    medians = []
    modes = []
    
    for i in columns:
        stds.append(np.std(data[i]))
        means.append(np.mean(data[i]))
        medians.append(np.median(data[i]))
        modes.append(stats.mode(data[i]))
        
    return stds, means, medians, modes
    
```

```python
# fit gaussians 
# indicate the mean
# indicate the std on each gaussian -- find a clever way of doing that (google is your friend)
# alternative option: plot all the means on one plot and see what happens
# note: think about the color scheme that is appropriate to what you want to show


# here i will make a piece of code that does this
# but only for one pdf
# I can then extend this to work for the whoel dataset
counts, bins, _ = plt.hist(data["pdf1"],
                          bins=50,
                          color="k",
                          histtype="step",
                          label="Data")

mu = np.mean(data["pdf1"])
sigma = np.std(data["pdf1"])

bin_width = bins[1] - bins[0]
area = np.sum(counts) * bin_width

gauss_fit = norm.pdf(bins, mu, sigma) * area

plt.plot(bins, gauss_fit, color="orange", label="Gaussian fit")

plt.vlines(mu, ymin=0, ymax=np.max(gauss_fit), color="orange")

xfill = np.arange(mu-sigma, mu+sigma, 0.01)
yfill = norm.pdf(xfill, mu, sigma) * area
plt.fill_between(xfill, yfill, 0, alpha=0.5, color="k")

plt.scatter([], [], label=f"$ \mu = {mu:.2f}$", color="white")
plt.scatter([], [], label=f"$ \sigma = {sigma:.2f}$", color="white")

plt.legend(frameon=False)
```

```python
def fit_gauss_and_plot(data, column):
    
    """
    add a doc string here!
    """
    
    counts, bins, _ = plt.hist(data,
                          bins=50,
                          color="k",
                          histtype="step",
                          label="Data")

    mu = np.mean(data)
    sigma = np.std(data)

    bin_width = bins[1] - bins[0]
    area = np.sum(counts) * bin_width

    gauss_fit = norm.pdf(bins, mu, sigma) * area

    plt.plot(bins, gauss_fit, color="orange", label="Gaussian fit")

    plt.vlines(mu, ymin=0, ymax=np.max(gauss_fit), color="orange")

    xfill = np.arange(mu-sigma, mu+sigma, 0.01)
    yfill = norm.pdf(xfill, mu, sigma) * area
    plt.fill_between(xfill, yfill, 0, alpha=0.5, color="k")

    plt.scatter([], [], label=f"$ \mu = {mu:.2f}$", color="white")
    plt.scatter([], [], label=f"$ \sigma = {sigma:.2f}$", color="white")
    
    plt.title(f"Gaussian fit for {column}")
    plt.legend(frameon=False)
    plt.savefig(f"gauss_fit_plot_{column}.pdf")
    plt.show()
    
    return
```
